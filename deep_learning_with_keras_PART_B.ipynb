```
import keras
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

import statistics
```

```
# Import data set
concrete_data = pd.read_csv('https://cocl.us/concrete_data')
concrete_data
```

```
# Name the features and label
predictors = ['Cement', 'Blast Furnace Slag', 'Fly Ash', 'Water', 'Superplasticizer', 'Coarse Aggregate', 'Fine Aggregate', 'Age']
target = ['Strength']
```

```
# Normalize the data
normalized_concrete_data = (concrete_data - concrete_data.min()) / (concrete_data.max() - concrete_data.min())
normalized_concrete_data
```

```
# Randomly split the data into a training and test sets by holding 30% of the data for testing
train, test = train_test_split(normalized_concrete_data, test_size=0.3)
```

```
# Build the neural net model
model = Sequential()
n_cols = concrete_data[predictors].shape[1]

# Single hidden layer with 10 nodes and ReLU activation function
model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
model.add(Dense(1))
```

```
mse_list = []

# Train the model 50 times and record the mean squared error
for i in range(50):
  # Fit the model using the adam optimizer and the mean squared error as the loss function
  model.compile(optimizer='adam', loss='mean_squared_error')
  model.fit(train[predictors], train[target], epochs=50)

  # Make predictions and record the MSE in a list
  predictions = model.predict(test[predictors])
  mse_list.append(mean_squared_error(test[target], predictions))

mse_list  
```

```
# Report the mean and the standard deviation of the mean squared errors.
mse_entries = len(mse_list)
mse_mean = statistics.mean(mse_list)
mse_stdev = statistics.stdev(mse_list)

print('Number of MSEs calculated = ' + str(mse_entries))
print('Mean of MSEs = ' + str(mse_mean))
print('Std dev of MSEs = ' + str(mse_stdev))
```
